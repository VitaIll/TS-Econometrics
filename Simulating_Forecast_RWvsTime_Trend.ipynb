```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from numpy import linalg as lin
from numpy import random as rnd

l=100
h=50
y1 = np.zeros(l)
y2 = np.zeros(l)
f1 = np.zeros(h)
f2 = np.zeros(h)
b1 = np.zeros(h)
b11 = np.zeros(h)
b2 = np.zeros(h)
b22 = np.zeros(h)

y1[0] = rnd.normal(0,1)
y2[0] = rnd.normal(0,1)
δ = 0.5
μ = 0.6
for i in range(1,l):
    y1[i] = δ + y1[i-1] + rnd.normal(0,1)
    y2[i] = i*μ + rnd.normal(0,1)
    if i>= l-h:
        f1[i-(l-h)] = y1[l-h]+(i-(l-h))*δ
        f2[i-(l-h)] = i*μ
        b1[i-(l-h)] = (i-(l-h))+y1[l-h]
        b11[i-(l-h)] = -(i-(l-h))+y1[l-h]
        b22[i-(l-h)] = 1+f2[i-(l-h)] 
        b2[i-(l-h)] = f2[i-(l-h)]-1

x1 = np.arange(0,l-h)
x2= np.arange(l-h,l)
fig, axs = plt.subplots(1,2,figsize=(15,4))
axs[0].plot(x1,y1[:l-h],label= "simulation log GDP")
axs[0].plot(x2,f1,label= "forecast")
axs[0].plot(x2,b1,color = 'g',label= "error band")
axs[0].plot(x2,b11,color = 'g')
axs[0].legend()
axs[0].set_title("predicting with random walk")


axs[1].plot(x1,y2[:l-h],label= "simulation log GDP")
axs[1].plot(x2,f2,label= "forecast")
axs[1].plot(x2,b2,color = 'g',label= "error band")
axs[1].plot(x2,b22,color = 'g')
axs[1].legend()
axs[1].set_title("predicting with time stationary trend")

plt.savefig("forc.jpg")
```
